```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_filename = "train.csv"
test_filename = "test.csv"
```

```python
train_load = pd.read_csv(train_filename)
test_load = pd.read_csv(test_filename)
```

```python
train_load
```

```python
train_load['source'] = 'train'
test_load['source'] = 'test'
```

```python
merged_data = train_load.append(test_load, ignore_index = True)
```

```python
#1)check for nulls (missing values)
merged_data.apply(lambda x: sum(x.isnull()))
```

```python
merged_data.apply(lambda x: len(x.unique()))
```

```python
#Filter categorical variables
categorical_columns = [x for x in merged_data.dtypes.index if merged_data.dtypes[x]=='object']
#Exclude ID cols and source:
categorical_columns = [x for x in categorical_columns if x not in ['Item_Identifier','Outlet_Identifier','source']]
#Print frequency of categories
for col in categorical_columns:
    print ('\nFrequency of Categories for varible %s'%col)
    print (merged_data[col].value_counts())
```

```python
merged_data.describe()
```

```python
#2) Inappropriate Values
#Determine average visibility of a product
visibility_avg = merged_data.pivot_table(values='Item_Visibility', index='Item_Identifier')
visibility_avg
```

```python

#Impute 0 values with mean visibility of that product:
item_visibility_zero = (merged_data['Item_Visibility'] == 0)

print('Number of 0 values initially: %d'%sum(item_visibility_zero))
merged_data.loc[item_visibility_zero,'Item_Visibility'] = merged_data.loc[item_visibility_zero,'Item_Identifier'].apply(lambda x: visibility_avg[x])
print('Number of 0 values after modification: %d'%sum(merged_data['Item_Visibility'] == 0))
```

```python
#3) Remove nd impute the bad data
#Determine the average weight per item:
item_avg_weight = merged_data.pivot_table(values='Item_Weight', index='Item_Identifier')
item_avg_weight
```

```python
#Get a boolean variable specifying missing Item_Weight values
item_weight_null = merged_data['Item_Weight'].isnull() 
item_weight_null
```

```python
#Impute data and check #missing values before and after imputation to confirm
print('Orignal #missing: %d'% sum(item_weight_null))
merged_data.loc[item_weight_null,'Item_Weight'] = merged_data.loc[item_weight_null,'Item_Identifier'].apply(lambda x: item_avg_weight[x])
print ('Final #missing: %d'% sum(merged_data['Item_Weight'].isnull()))
```

```python
#check for nulls (missing values)
merged_data.apply(lambda x: sum(x.isnull()))
```

```python
#Import mode function:
from scipy.stats import mode

#Determing the mode for each
outlet_size_mode = merged_data.pivot_table(values='Outlet_Size', columns='Outlet_Type',aggfunc=(lambda x:mode(x.astype('str')).mode[0]) )
print('Mode for each Outlet_Type:')
print(outlet_size_mode)

#Get a boolean variable specifying missing outlet_size values
outlet_size_null = merged_data['Outlet_Size'].isnull() 

#Impute data and check #missing values before and after imputation to confirm
print('\nOriginal missing: %d'% sum(outlet_size_null))
merged_data.loc[outlet_size_null,'Outlet_Size'] = merged_data.loc[outlet_size_null,'Outlet_Type'].apply(lambda x: outlet_size_mode[x])
print('\nAfter Cleaning: %d'%sum(merged_data['Outlet_Size'].isnull()))
```

```python
clean_train_data = merged_data.ix[(merged_data['source'] == 'train')]
clean_train_data = clean_train_data.drop(['source'], axis = 1)
clean_test_data = merged_data.ix[(merged_data['source'] == 'test')]
clean_test_data = clean_test_data.drop(['source'], axis = 1)
```

```python
merged_data.to_csv("merged_clean.csv", sep = ',')
clean_train_data.to_csv("train_clean.csv", sep = ',')
clean_test_data.to_csv("test_clean.csv", sep = ',')
```

```python
#Anomalies / Outliers
import numpy as np

def outliers_z_score(ys):
    threshold = 3
    mean_y = np.mean(ys)
    stdev_y = np.std(ys)
    z_scores = [(y - mean_y) / stdev_y for y in ys]
    return np.where(np.abs(z_scores) > threshold)
```

```python

plt.hist(outliers_z_score(train_load['Item_Outlet_Sales']), bins=np.arange(0, 5000, 200))

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Diagramle')
plt.grid(True)
```

```python
from scipy import stats
fit = stats.norm.pdf(outliers_z_score(train_load['Item_Outlet_Sales']), np.mean(train_load['Item_Outlet_Sales']), np.std(train_load['Item_Outlet_Sales']))  #this is a fitting indeed

plt.plot(outliers_z_score(train_load['Item_Outlet_Sales']),fit,'-o')

#plt.hist(outliers_z_score(train_load['Item_Outlet_Sales']),normed=True)      #use this to draw histogram of your data

plt.show()             
```

```python
#Get the first two characters of ID:
merged_data['Item_Type_Combined'] = merged_data['Item_Identifier'].apply(lambda x: x[0:2])
#Rename them to more intuitive categories:
merged_data['Item_Type_Combined'] = merged_data['Item_Type_Combined'].map({'FD':'Food',
                                                             'NC':'Non-Consumable',
                                                             'DR':'Drinks'})
merged_data['Item_Type_Combined'].value_counts()
```

```python
#Change categories of low fat:
print('Original Categories:')
print(merged_data['Item_Fat_Content'].value_counts())

print('\nModified Categories:')
merged_data['Item_Fat_Content'] = merged_data['Item_Fat_Content'].replace({'LF':'Low Fat',
                                                             'reg':'Regular',
                                                             'low fat':'Low Fat'})
print(merged_data['Item_Fat_Content'].value_counts())
```

```python
#Mark non-consumables as separate category in low_fat:
merged_data.loc[merged_data['Item_Type_Combined']=="Non-Consumable",'Item_Fat_Content'] = "Non-Edible"
merged_data['Item_Fat_Content'].value_counts()
```

```python
#one-hot encoding - creating dummy variables
#Import library:
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
#New variable for outlet
merged_data['Outlet'] = le.fit_transform(merged_data['Outlet_Identifier'])
var_mod = ['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Item_Type_Combined','Outlet_Type','Outlet']
le = LabelEncoder()
for i in var_mod:
    merged_data[i] = le.fit_transform(merged_data[i])
```

```python
#One Hot Coding:
merged_data = pd.get_dummies(merged_data, columns=['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Outlet_Type',
                              'Item_Type_Combined','Outlet'])
```

```python
merged_data
```

```python
#Drop the columns which have been converted to different types:
merged_data.drop(['Item_Type','Outlet_Establishment_Year'],axis=1,inplace=True)

#Divide into test and train:
model_train = merged_data.loc[merged_data['source']=="train"]
model_test = merged_data.loc[merged_data['source']=="test"]

#Drop unnecessary columns:
model_test.drop(['source'],axis=1,inplace=True)
model_train.drop(['source'],axis=1,inplace=True)
```

```python
#Export files as modified versions:
model_train.to_csv("model_train.csv",index=False)
model_test.to_csv("model_test.csv",index=False)
```

```python
ytrain = model_train['Item_Outlet_Sales']
xtrain = model_train.drop(['Item_Outlet_Sales'])
ytest = model_test['Item_Outlet_Sales']
xtest = model_test.drop(['Item_Outlet_Sales'], axis=1,inplace=True)
```

```python
#random forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import *
from sklearn.model_selection import cross_val_score

max_depth = 5
regr_rf = RandomForestRegressor(max_depth=max_depth, random_state=14)
regr_rf.fit(xtrain, ytrain)
print('Random Forest -> Fit successful')
#y_rf = regr_rf.predict(xtest)
score=cross_val_score(regr_rf,xtrain,ytrain).mean()
#print(y_rf)
print('Random Forest -> Score : ',score)
```
